#### fit_on_texts() gives the following attributes in the output as given [here](https://faroit.github.io/keras-docs/1.2.2/preprocessing/text/).

* **word_counts:** dictionary mapping words (str) to the number of times they appeared on during fit. Only set after fit_on_texts was called.

* **word_docs:** dictionary mapping words (str) to the number of documents/texts they appeared on during fit. Only set after fit_on_texts was called.

* **word_index:** dictionary mapping words (str) to their rank/index (int). Only set after fit_on_texts was called.

* **document_count:** int. Number of documents (texts/sequences) the tokenizer was trained on. Only set after fit_on_texts or fit_on_sequences was called.

````python
### initializing the placeholder variable
data = np.zeros((len(texts), MAX_SENTS, MAX_SENT_LENGTH), dtype='int32')

### word index encoding
for i, sentences in enumerate(articles):
    for j, sent in enumerate(sentences):
        if j < MAX_SENTS:
            wordTokens = text_to_word_sequence(sent)
            k = 0
            for _, word in enumerate(wordTokens):
                try:
                    if k < MAX_SENT_LENGTH and tk.word_index[word] < MAX_NB_WORDS:
                      data[i, j, k] = tk.word_index[word]
                      k = k + 1
                except:
                      #  print(word)
                        pass
````

### Create embedding matrix with the glove embeddings


Run the below code to create embedding_matrix which has all the words and their glove embedding if present in glove word list.

````python
len(tk.word_index)
````

````python
vocab_size = len(tk.word_index)
````

````python
# load the whole embedding into memory
embeddings_index = dict()
f = open('./glove.6B.100d.txt')
for line in f:
	values = line.split()
	word = values[0]
	coefs = np.asarray(values[1:], dtype='float32')
	embeddings_index[word] = coefs
f.close()
print('Loaded %s word vectors.' % len(embeddings_index))

# create a weight matrix for words in training docs
embedding_matrix = np.zeros((vocab_size, 100))


for word, i in tk.word_index.items():
	embedding_vector = embeddings_index.get(word)
	if embedding_vector is not None:
		embedding_matrix[i] = embedding_vector
````

# Try the sequential model approach and report the accuracy score. [10 marks]  

### Import layers from Keras to build the model

````python
from keras.models import Sequential, Model
from keras.layers import Dropout, Activation
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D, Concatenate, Input, Add
from keras.layers import Bidirectional, TimeDistributed
from keras.callbacks import EarlyStopping
````

### Model

````python
### Model only for Article Body

token_num = MAX_NB_WORDS
data_length = MAX_SENT_LENGTH
output_length = 100 
lstm_dim = 256   ## try with 128 as well
input_length = MAX_SENT_LENGTH * MAX_SENT_LENGTH


model = Sequential()
model.add(Embedding(output_dim = output_length, input_dim = len(tk.word_counts), input_length = 400, weights=[embedding_matrix]))
model.add(Dropout(0.2))

model.add(Bidirectional(LSTM(lstm_dim), merge_mode = 'sum'))
model.add(Dropout(0.2))

model.add(Dense(units = 256, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(units = 4, activation = 'softmax'))
print(model.summary())
````

### Compile and fit the model

````python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
````

````python
# cb = EarlyStopping(monitor = 'val_loss', patience = 5, verbose = 2)
````

````python
# Ran the model only with Aricle body data to compile and check the initial accuracy
train_history = model.fit(x = x_train_re,
                         y = y_train,
                         validation_data = (x_val_re, y_val),
                         epochs = 2,
                         batch_size = 100,
                         verbose = 1)
````

### Final model concat with Header along with Body

````python
token_num = len(tk.word_counts)  ### MAX_NB_WORDS
output_length = 100 
lstm_dim = 256   ## try with 128 as well
input_length_body = MAX_SENTS * MAX_SENT_LENGTH
input_length_header = MAX_SENTS_HEADING * MAX_SENT_LENGTH

model_body = Sequential()
model_body.add(Embedding(output_dim = output_length, input_dim = token_num, input_length = input_length_body, weights=[embedding_matrix]))
model_body.add(Dropout(0.2))
model_body.add(Bidirectional(LSTM(lstm_dim), merge_mode = 'sum'))
model_body.add(Dropout(0.2))
model_body.add(Dense(units = 256, activation = 'relu'))
model_body.add(Dropout(0.2))

model_header = Sequential()
model_header.add(Embedding(output_dim = output_length, input_dim = token_num, input_length = input_length_header, weights=[embedding_matrix]))
model_header.add(Dropout(0.2))
model_header.add(Bidirectional(LSTM(lstm_dim), merge_mode = 'sum'))
model_header.add(Dropout(0.2))
model_header.add(Dense(units = 256, activation = 'relu'))
model_header.add(Dropout(0.2))

# final_model = Sequential()

# final_model.add(Dense(units = 1, input_shape=(256,)))
# final_model.add(Concatenate([model_body, model_header]))
merged_output = Add()([model_body.output, model_header.output])
merged_output = Dense(units = 4, activation = 'softmax')(merged_output)

final_model = Model([model_body.input, model_header.input], merged_output)

print(final_model.summary())
````

````python
final_model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
````

````python
train_history = final_model.fit(x = [x_train_re, x_heading_train_re],
                         y = y_train,
                         validation_data = ([x_val_re, x_heading_val_re], y_val),
                         epochs = 1,
                         batch_size = 100,
                         verbose = 1)
````
